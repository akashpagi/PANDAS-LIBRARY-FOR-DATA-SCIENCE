#### What is Pandas DataFrame?
- A Pandas DataFrame is a 2 dimensional data structure, like a 2 dimensional array, or a table with rows and columns.
- Used for data manipulation, analysis, and cleaning, featuring labeled rows and columns, support for various data types, and flexibility for adding, removing, and transforming data.
- Pandas DataFrames are widely used in data science and analysis tasks for handling structured data efficiently.
- **Syntax: pandas.DataFrame(data=None, index=None, columns=None, dtype=None, copy=None)**
    - **data:** ndarray (structured or homogeneous), Iterable, dict, or DataFrame
    - **index:** Index or array-like
    - **columns:** Index or array-like
    - **dtype:** dtype, default None
    - **copy:** bool or None, default None

**rename() method:**
- rename() method is used to rename any index, column or row. Renaming of column can also be done by dataframe.columns = [#list]
- **Syntax: DataFrame.rename(mapper,index,columns,axis,copy, inplace,level)**
    - **mapper, index and columns:** Dictionary value, key refers to the old name and value refers to new name. Only one of these parameters can be used at once.
    - **axis:** int or string value, 0/’row’ for Rows and 1/’columns’ for Columns.
    - **copy:** Copies underlying data if True.
    - **inplace:** Makes changes in original Data Frame if True.
    - **level:** Used to specify level in case data frame is having multiple level index.



```python
students.rename(columns={'marks':'%','package':'lpa'}, inplace=True)
print(students)
```

#### Selecting rows from a Pandas DataFrame

**Searching using index position**

**iloc[] method**
- We can extract the rows by using an imaginary index position which is not visible in the DataFrame.
- Allowed inputs for .loc[] are Integer value, List or array of integers.Slice object with ints, boolean array and a callable function with one argument that can be the calling Series or the DataFrame. It returns valid outputs for indexing.
- **Syntax: DataFrame.iloc[index_position]**

```python
# Single row (horizontal)
movies.iloc[0]
```

```python
# Multiple rows using fancy indexing
movies.iloc[[1,5,6]]
```

```python
# Multiple rows using slicing
movies.iloc[:11:2]
```

**Searching using index label**

**loc[] method**

- The DataFrame.loc[] is used to retrieve the group of rows and columns by labels or a boolean array in the DataFrame. 
- It takes only index labels, and if it exists in the caller DataFrame, it returns the rows, columns, or DataFrame.
- The DataFrame.loc[] is a label based but may use with the boolean array.
- The allowed inputs for .loc[] are:
    - Single label, e.g.,7 or a. Here, 7 is interpreted as the label of the index.
    - List or array of labels, e.g. ['x', 'y', 'z'].
    - Slice object with labels, e.g. 'x':'f'.
    - A boolean array of the same length. e.g. [True, True, False].
    - callable function with one argument.
- **Syntax: DataFrame.loc[index_label]**

```python
dictionary = {
    'name':['goku','eren','luffy','light','king'],
    'iq':[100,95,14,0,0],
    'marks':[107,87,16,0,0],
    'package':[89,78,12,0,0],
}
students = pd.DataFrame(dictionary)
students.set_index('name', inplace=True)
students
```

```python
# Single index 
students.loc['goku']
```

```python
# Multiple index
students.loc[['goku','eren','light']]
```
